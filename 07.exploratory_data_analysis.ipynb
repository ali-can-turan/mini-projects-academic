# Task 1 - Visualize individual feature patterns

### Continuous valued features
Generate regression plots for each of the parameters "CPU_frequency", "Screen_Size_inch" and "Weight_pounds" against "Price". Also, print the value of correlation of each feature with "Price".


```python
# Write your code below and press Shift+Enter to execute
# CPU_frequency plot
print(stats.pearsonr(df["CPU_frequency"], df["Price"]))
sns.regplot(x="CPU_frequency",
            y="Price",
            data=df,
            y_jitter=True,
            scatter_kws={"color":"r"}
           );
```

```python
# Write your code below and press Shift+Enter to execute
# Screen_Size_inch plot
print(stats.pearsonr(df["Screen_Size_inch"], df["Price"]))
sns.regplot(x="Screen_Size_inch",
            y="Price",
            data=df,
            y_jitter=True,
            marker="d",
            scatter_kws={"color":"orange"}
           );
```

```python
# Write your code below and press Shift+Enter to execute
# Weight_pounds plot
print(stats.pearsonr(df["Weight_pounds"], df["Price"]))
sns.regplot(x="Weight_pounds",
            y="Price",
            data=df,
            y_jitter=True,
            marker="*",
            scatter_kws={"color":"pink"}
           );
```

Interpretation: "CPU_frequency" has a 36% positive correlation with the price of the laptops. The other two parameters have weak correlation with price.

# Task 3 - GroupBy and Pivot Tables

Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.


```python
# Write your code below and press Shift+Enter to execute
# Create the group
df_group = df[["GPU", "CPU_core", "Price"]]
agg = df_group.groupby(["GPU", "CPU_core"], as_index=False).mean()
agg
```

```python
# Write your code below and press Shift+Enter to execute
# Create the Pivot table
pivot = agg.pivot_table(columns="CPU_core", values="Price", index="GPU")
pivot
```

```python
# Write your code below and press Shift+Enter to execute
# Create the Plot
sns.heatmap(data=pivot,
            mask= pivot < 800,
           center=1500,
           square=True,
           annot=True,
           fmt="1.2f",
           cbar=True,
           cbar_kws={"label":"scale"},
           ec="1.0",
           cmap="viridis",           
          );
```

# Task 4 - Pearson Correlation and p-values

Use the `scipy.stats.pearsonr()` function to evaluate the Pearson Coefficient and the p-values for each parameter tested above. This will help you determine the parameters most likely to have a strong effect on the price of the laptops.


```python
# Write your code below and press Shift+Enter to execute
for column in df:
    if pd.api.types.is_numeric_dtype(df[column]) and column != "Price" and abs(stats.pearsonr(df[column], df["Price"])[0]) > 0.3:
        print(f"Stats for {column} and Price:")
        print(stats.pearsonr(df[column], df["Price"]),"\n")
```
